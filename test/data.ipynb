```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from data.generators import *
```

### 1D Convex Data Generation

```python
X_train, X_val, y_train, y_val = generate_1d_convex(n_samples=1000)
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Validation set size: {X_val.shape[0]} samples")
```

Visualization for the 1D Convex Data Generation.

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_train, y_train, label='Train Data', color='blue', alpha=0.5)
plt.scatter(X_val, y_val, label='Validation Data', color='green', alpha=0.5)
plt.title('1D Convex Function: Training and Validation Data')
plt.xlabel('X')
plt.ylabel('f(X) = XÂ²')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("1D Convex Function - Summary Statistics:")
print(f"X_train: min={X_train.min()}, max={X_train.max()}, mean={X_train.mean():.4f}")
print(f"y_train: min={y_train.min()}, max={y_train.max()}, mean={y_train.mean():.4f}")
print(f"X_val: min={X_val.min()}, max={X_val.max()}, mean={X_val.mean():.4f}")
print(f"y_val: min={y_val.min()}, max={y_val.max()}, mean={y_val.mean():.4f}")
```

### 2D Convex Data Generation

Since we are working on 2D, 10000 samples should be enough to get a good approximation of the function

```python
n_samples = 10000
X_train, X_val, y_train, y_val = generate_2d_convex(n_samples=n_samples)
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Validation set size: {X_val.shape[0]} samples")
print(f"Input Dimensions: {X_train.shape[1]}")
```

Scatter plot for the 2D Convex Data Generation.

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis', alpha=0.5)
plt.title('2D Convex Function: Training Data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.colorbar(label='f(X) = X1Â² + X2Â²')
plt.grid(True)
plt.show()
```

```python
print("2D Convex Function - Summary Statistics:")
print(f"X_train: min={X_train.min()}, max={X_train.max()}, mean={X_train.mean():.4f}")
print(f"y_train: min={y_train.min()}, max={y_train.max()}, mean={y_train.mean():.4f}")
print(f"X_val: min={X_val.min()}, max={X_val.max()}, mean={X_val.mean():.4f}")
print(f"y_val: min={y_val.min()}, max={y_val.max()}, mean={y_val.mean():.4f}")
```
